```python
import pandas as pd
```

```python
data = pd.read_csv('Hotel Reservations.csv')

data
```

```python
data['market_segment_type'].unique()
```

```python
data.info()
```

```python
data.drop('Booking_ID', axis=1, inplace=True)
```

```python
data.info()
```

```python
data['type_of_meal_plan'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
for col in data.columns:
  if data[col].dtype == 'object':
    data[col] = encoder.fit_transform(data[col])
```

```python
data.info()
```

```python
x = data.drop(columns=['booking_status', 'arrival_year'], axis=1)
y = data['booking_status']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
from pycaret.classification import *
```

```python
classification_model = setup(data=data, target='booking_status')

best_classification_model = compare_models()
```

```python
classification = create_model('rf')

classification.fit(x_train, y_train)
```

```python
import pickle

with open('hotel_reservation_model.pkl', 'wb') as file:
    pickle.dump(classification, file)
```
